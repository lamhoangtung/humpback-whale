```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from config import *
```

Đọc `train.csv` thành dataframe

```python
train_raw = pd.read_csv(train_csv)
train_raw.shape
```

Loại class `new_whale` ra khỏi train set

```python
train_raw = train_raw.loc[train_raw['Id'] != 'new_whale']
train_raw.shape
```

```python
train_raw.head()
```

Cách chia validation:
- Với class chỉ có một sample: sample nằm trong cả train và test
- Với class có ít hơn 10 sample: 1 sample test, train phần còn lại
- Với các class còn lại: chia 90% train và 10% test

```python
final_train = pd.DataFrame(columns=['Image', 'Id'])
final_test = pd.DataFrame(columns=['Image', 'Id'])

all_class = train_raw.groupby('Id')
for each_class in tqdm(all_class):
    num_of_samples = each_class[1].shape[0]
    if num_of_samples == 1:
        final_train = final_train.append(each_class[1], ignore_index=True)
        final_test = final_test.append(each_class[1], ignore_index=True)
    elif num_of_samples <= 10:
        test_sample = each_class[1].sample(n=1, random_state=200)
        train_sample = each_class[1].drop(test_sample.index)
        final_train = final_train.append(train_sample, ignore_index=True, sort=False)
        final_test = final_test.append(test_sample, ignore_index=True, sort=False)
    else:
        test_sample = each_class[1].sample(frac=0.1, random_state=200)
        train_sample = each_class[1].drop(test_sample.index)
        final_train = final_train.append(train_sample, ignore_index=True, sort=False)
        final_test = final_test.append(test_sample, ignore_index=True, sort=False)
```
